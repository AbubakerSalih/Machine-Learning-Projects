```python
  import pandas as pd
  from sklearn.model_selection import train_test_split
  import numpy as np
  import matplotlib.pyplot as plt
  import seaborn as sns
  from xgboost import XGBRegressor
  from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
df= pd.read_csv("/BostonHousing.csv")
```

```python
# since I'm doing regression I'm going to drop the CAT. MEDV column
df=df.drop("CAT. MEDV",axis=1)
df.head()
```

```python
corr=df.corr()["MEDV"].sort_values(ascending=False)
corr
```

```python
corrlation = df.corr()
plt.figure(figsize=(8,8))
sns.heatmap(corrlation,cbar=True,annot=True,square=True,fmt=".2f",annot_kws={"size":8})
```

```python
X=df.drop("MEDV",axis=1)
Y= df["MEDV"]
print(X,Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
print(X_train.shape,X_test.shape)
```

```python
model=XGBRegressor()
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python
# R2
r2 = r2_score(Y_pred,Y_test)
# RMSE
RMSE = np.sqrt(mean_squared_error(Y_pred,Y_test,))
#MAE
MAE = mean_absolute_error(Y_pred,Y_test)

print("R2 is ", r2)
print("RMSE is ",RMSE)
print("MAE is ",MAE)
```

```python
#Plotting
plt.figure(figsize=(6,6))
plt.scatter(Y_test,Y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```

```python
# Adding a reference line y = x
plt.figure(figsize=(6,6))
plt.scatter(Y_test, Y_pred)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', linestyle='--')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```
