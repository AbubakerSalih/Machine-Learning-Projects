```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rupakroy/sonarcsv")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv("/kaggle/input/sonarcsv/sonar.csv",header=None)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

```python
x= df.drop(60,axis=1)
y=df[60]
print(x)
print(y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)

# stratify ensures that the train and test sets have the same proportion of classes as the original dataset y
```

```python
print(x.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

#training the model

```python
# training the model
model.fit(X_train,Y_train)
```

#Evaluation

```python
#accuracy on training data
X_train_predeiction=model.predict(X_train)
training_data_accuracy=accuracy_score(Y_train,X_train_predeiction)
```

```python
print(f"Accuracy on training data is: {training_data_accuracy}")
```

```python
#accuracy on testing data
X_test_predeiction=model.predict(X_test)
testing_data_accuracy=accuracy_score(Y_test,X_test_predeiction)
```

```python
print(f"Accuracy on testing data is: {testing_data_accuracy}")
```
